## Time Complexity of Brute Force, Grover and Shor

This notebook compares the time complexity (number of operations) required to find out a certain key of n bits length using different methods

```python
# Imports
from math import log, log2
import matplotlib.pyplot as plt
```

```python
# Set up data, we use a log2 scale for all algorithms
number_of_bits = list(range(1, 256))
brute_force = []
grover = []
shor = []

# Calculate number of operations
for x in number_of_bits:
    # Number of Brute Force operations in a binary log scale
    # log2(2**x) == x
    brute_force.append(x)
    # Number of operations with Grover in a binary log scale
    # log2(sqrt(2**x)) == 0.5*x
    grover.append(0.5*x)
    # Number of operations with Shor in a binary log scale
    # log2(log(2**x)**3) == 3 * log2(x * log(2))
    shor.append(3 * log2(x * log(2)))
```

```python
# Visualize results
# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(number_of_bits, brute_force, label="Brute Force Attack 2**x", color='blue')
plt.plot(number_of_bits, grover, label="Grover's Search sqrt(2**x)", color='green')
plt.plot(number_of_bits, shor, label="Shor's Algorithm log(2**x)**3", color='red')

# Add labels and title
plt.xlabel('Bit length (x)')
plt.ylabel('Time Complexity (log2 scale)')
plt.title("Brute Force vs Grover vs Shor")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
